```python
import os
import fileinput
import matplotlib.pyplot as plt
from nltk.stem import WordNetLemmatizer 
from nltk.stem import PorterStemmer
```

```python
files = [f"documents/{ff}" for ff in os.listdir("documents")]
```

```python
def next_word(files):
    inputs = fileinput.input(files)
    for line in inputs:
        tokens = line.split(" ")
        for token in tokens:
            yield token
    inputs.close()
```

# Palabras como tokens

```python
token_stream = next_word(files)
```

```python
token_count = 0
vocabulary = set()
x = []
y = []
for token in token_stream:
    token_count += 1
    vocabulary.add(token)
    x.append(token_count)
    y.append(len(vocabulary))
    
```

```python
plt.plot(x, y)
plt.xlabel('# Tokens')
plt.ylabel('Tamaño del vocabulario')
plt.show()
```

# Lemmas como tokens

```python
token_stream = next_word(files)
```

```python
token_count = 0
vocabulary = set()
x = []
y = []
lemmatizer = WordNetLemmatizer() 
for token in token_stream:
    token_count += 1
    lemma = lemmatizer.lemmatize(token)
    vocabulary.add(lemma)
    x.append(token_count)
    y.append(len(vocabulary))
    
```

```python
plt.plot(x, y)
plt.xlabel('# Lemmas')
plt.ylabel('Tamaño del vocabulario')
plt.show()
```

# Stems como tokens

```python
token_stream = next_word(files)
```

```python
token_count = 0
vocabulary = set()
x = []
y = []
stemmer = PorterStemmer() 
for token in token_stream:
    token_count += 1
    stem = stemmer.stem(token)
    vocabulary.add(stem)
    x.append(token_count)
    y.append(len(vocabulary))
    
```

```python
plt.plot(x, y)
plt.xlabel('# Stems')
plt.ylabel('Tamaño del vocabulario')
plt.show()
```
